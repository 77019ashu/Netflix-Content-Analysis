```python
netflix_data['director'].fillna('Unknown', inplace=True)
netflix_data['cast'].fillna('Unknown', inplace=True)
netflix_data
```

```python
netflix_data['country'].fillna('Unknown', inplace=True)
netflix_data['date_added'].fillna('Unknown', inplace=True)
netflix_data['rating'].fillna('Unrated', inplace=True)
netflix_data
```

```python
netflix_data['date_added'] = pd.to_datetime(netflix_data['date_added'], errors='coerce')
netflix_data['date_added']
```

```python
def extract_duration(duration):
    if isinstance(duration, float) and np.isnan(duration):
        return np.nan

    if 'Season' in duration:
        return int(duration.split(' ')[0])
    else:
        return int(duration.replace(' min', ''))

import numpy as np

netflix_data['duration'] = netflix_data['duration'].apply(extract_duration)
netflix_data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
plt.figure(figsize=(8,6))
sns.countplot(x='type', data=netflix_data, palette="coolwarm")
plt.title('Distribution of Movies vs. TV Shows on Netflix', fontsize=14)
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,6))
top_genres = netflix_data['listed_in'].str.split(', ').explode().value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, palette="viridis")
plt.title('Top 10 Genres on Netflix', fontsize=14)
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

```python
plt.figure(figsize=(10,6))
top_countries = netflix_data['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette="magma")
plt.title('Top 10 Countries Producing Netflix Content', fontsize=14)
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='rating', data=netflix_data, order=netflix_data['rating'].value_counts().index, palette="cubehelix")
plt.title('Distribution of Age Ratings on Netflix', fontsize=14)
plt.xlabel('Count')
plt.ylabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(netflix_data['release_year'], bins=30, kde=True, color="coral")

plt.title('Distribution of Release Years on Netflix', fontsize=14)

plt.xlabel('Release Year')

plt.ylabel('Number of Titles')

plt.show()
```

```python
plt.figure(figsize=(10,6))

top_directors = netflix_data['director'].value_counts().head(10)

sns.barplot(x=top_directors.values, y=top_directors.index, palette="plasma")

plt.title('Top 10 Directors on Netflix (by number of titles)', fontsize=14)

plt.xlabel('Number of Titles')

plt.ylabel('Director')

plt.show()
```

```python
netflix_data_filtered = netflix_data[netflix_data['date_added'].notna()]
netflix_data_filtered['month_added'] = netflix_data_filtered['date_added'].dt.month
plt.figure(figsize=(10,6))
sns.countplot(x='month_added', data=netflix_data_filtered, palette="Blues_d", order=range(1, 13))
plt.title('Content Added by Month on Netflix', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Number of Titles Added')
plt.show()
```
